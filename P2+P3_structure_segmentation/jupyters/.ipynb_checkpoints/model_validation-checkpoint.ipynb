## Cell distance mask

```python
test = np.load("./ide_folder/output/cell_seg/multiclass/2724_part_2.npy")
```

```python
test_1 = np.where(test == 1, 1, 0)
test_2 = np.where(test == 2, 1, 0)
test_3 = np.where(test == 3, 1, 0)
```

```python
np.unique(cv2.distanceTransform(np.where(test_1 == 1, 0, 1).astype(np.uint8), cv2.DIST_L2, 3))
```

```python
dst = cv2.distanceTransform(np.where(test_3 == 1, 0, 1).astype(np.uint8), cv2.DIST_L2, 3)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(np.where(dst > 255, 255, dst))
```
